#Data Cleaning and Processing Part 1 (handling missing values)

```python
# dropping unnecessary columns
df= df.drop(['web-scraper-order', 'web-scraper-start-url', 'all', 'all-href','page', 'page-href',"review_star"
],axis=1)
```

```python
df.isnull().sum()
```

```python
df[df['name'].isnull()]
```

Since every null "name" column contains null values in other columns, so dropping it.

```python
df.dropna(subset=['name'], inplace=True)
```

```python
df[df['product_details'].isnull()]
```

Since product details and ingredients are the two most important factors applying BERT model, so dropping the rows where both the columns are null.

```python
df = df.dropna(subset=['product_details'], how='all')
```

```python
df[df['ingredients'].isnull() & df['size'].isnull() ]
```

Then dropping all the rows where the "ingredient" and "size" columns have null values. The reason is that chances are high that those are skincare tools than a product since tools do not have any size and ingredients.



```python
df = df.dropna(subset=['size', 'ingredients'], how='all')
```

```python
df.isnull().sum()
```

```python
df[df['size'].isnull()]
```

```python
# drpping the rows where "size" column has null values
df.dropna(subset=['size'], inplace=True)
```

```python
df.isnull().sum()
```

```python
df[df['brand'].isnull()]
```

```python
# put the brand name from the product's "name" column 
df["brand"].fillna("Burt’s Bees", inplace = True)
```

I will not drop any ingredient column since some product details include ingredients. so I will Fill the null value with N/A

```python
df["ingredients"].fillna("N/A", inplace = True)
```

```python
df.isnull().sum()
```

All null values are handled

#Data Cleaning and Processing Part 2 (extracting and adding new column)

```python
# converting dataframe value in to lowercase 
df = df.apply(lambda x: x.astype(str).str.lower())
```

```python
# extracted the product amount number from "size" column and created a new column
df["product_amount"] = df["size"].str.extract(r'(\d+.\d+|\d+)')
```

```python
# extracted the measurement unit from "size" column and created a new column
df["measurement_unit"] = df["size"].str.extract(r'(ml|g|fl. oz|oz|fl.oz)')
```

```python
df
```

```python
df.isnull().sum()
```

```python
df[df['product_amount'].isnull()]
```

```python
df.dropna(subset=['product_amount'], inplace=True)
```

```python
df[df['measurement_unit'].isnull()]
```

```python
df.dropna(subset=['measurement_unit'], inplace=True)
```

Dropped all the null valued "product_amount" and "measurement_unit" columns since they are skincare tools rather than a skin products.

```python
df.isnull().sum()
```

No more null value.

```python
# checking variety in "size" column to understand what should be removed
marks_list = df['size'].tolist()
marks_list
```

As we can see, some entry has multiple products. We have identified that products that are multiple in number, contain strings like "x", "pairs", and "pack", etc. So we will remove those from the "name" and "size" columns since we want to work with an only single product.

```python
 df = df[df["size"].str.contains("x")==False]
df = df[df["size"].str.contains("X")==False]
df = df[df["size"].str.contains("pairs")==False]
df = df[df["size"].str.contains("sachet")==False]
df = df[df["size"].str.contains(":")==False]
df = df[df["size"].str.contains(",")==False]
df = df[df["size"].str.contains("-")==False]
```

```python
df = df[df["name"].str.contains("x")==False]
df = df[df["name"].str.contains("roller")==False]
df = df[df["name"].str.contains("pack")==False]
df = df[df["name"].str.contains("pairs")==False]
df = df[df["name"].str.contains("sachet")==False]
```

```python
df
```

```python
df.dtypes
```


The datatype should be changed for the numerical columns; "product_amount", "price".

```python
df["product_amount"]= df['product_amount'].astype(dtype ='float64')
```

Before changing the "price" column data type, we need to do some cleaning like removing the dollar sign.

```python
df["price"]=df["price"].str.strip("$ , msrp: $ ")
```

```python
df["price"]=df["price"].str.replace(',', '')
```

```python
df["price"]= df['price'].astype(dtype ='float64')
```

Now we will convert all the product amount in a single measurement unit. FOllowing are the convertion measurement. 

- 1 oz= 28.3495 gram
- 1 ml = 1 gram
- 1 fl.oz = 29.5735 ml= 29.5735 gram

we will convert every product amount into gram.

```python
df.measurement_unit.unique()
```

```python
df["measurement_unit"]=df['measurement_unit'].replace({'fl. oz': 'fl.oz'}).replace({'fl oz': 'fl.oz'})
```

```python
df.loc[df.measurement_unit == "oz","product_amount"] = df.product_amount * 28.3495
df.loc[df.measurement_unit == "fl.oz","product_amount"] = df.product_amount * 29.5735
```

```python
df["measurement_unit"]=df['measurement_unit'].replace({'fl.oz': 'gram'}).replace({'ml': 'gram'}).replace({'oz': 'gram'}).replace({'g': 'gram'})
```

```python
df.measurement_unit.unique()
```

```python
# rounding up the value
df['product_amount']=df['product_amount'].apply(lambda x:round(x,2))
```


Now we would like to add a new column "product_type" to check what type of products are these. So we will try to get the product type from the "name" and "product_details" columns. 

We have identified what type of word refer different product from reviewing the dataset and found the keywords

```python
df["product_details"]=df["product_details"].str.lower()
```

```python
List1 = ['moisturizer',"moisturiser","booster",'mist', 'tonic', "essence",'masque',"retinol","moisturising","wash","micellar","cleansing","facial wash","eye","lip","emulsion","cleanser","crème","treatment","exfoliator","exfoliant","toner","balm","cream","mask","oil","spf","serum","lotion"]
df["type1"] = df['name'].str.extract("(" + "|".join(List1) +")", expand=False)
df
```

```python
df["type1"].isnull().sum()
```

```python
List1 = ['moisturizer',"booster","moisturiser","essence",'masque',"retinol","wash","lip","micellar","facial wash","cleanser","crème","eye","treatment","scrub","face mist","facial mist","exfoliator","cleansing","exfoliant","toner","balm","eye cream","cream","mask","spf","serum","lotion"]
df["type2"] = df['product_details'].str.extract("(" + "|".join(List1) +")", expand=False)
df.tail(20)
```

```python
df["type2"].isnull().sum()
```

Since both the "name" and "product_details" column could not identify all the products, we merge them by filling the N/A value of "type2" by "type1" value.

```python
 df["product_type"] = df["type1"].fillna(df["type2"])
```

```python
df
```

```python
df["product_type"].isnull().sum()
```

```python
df.dropna(subset=['product_type'], inplace=True)
```

```python
# dropping the unnecessary column
df= df.drop(['size', 'type1',"type2"], axis=1)
```

```python
df
```

From researching on internet, we tried to capture the standard size of products. we have found that, small/travel size products are under or equal to 30 gram. in general, if a product is more than 300 gram, it is considered as super in size. Everything in middle in standard in size.

```python
df.loc[df['product_amount'] <= 30, 'product_size'] = 'small' 
df.loc[df['product_amount'] > 30, 'product_size'] = 'standard'
df.loc[df['product_amount'] > 300, 'product_size'] = 'large'
```

```python
df
```

```python
df['product_type'].unique() 
```

Previously we have used different keywords to get our product type. Let's uniform them in a single type below-

```python
df["product_type"]=df['product_type'].replace({'cream': 'cream/lotion/moisturizer'}).replace({'lotion': 'cream/lotion/moisturizer'}).replace({'moisturiser': 'cream/lotion/moisturizer'}).replace({'moisturizer': 'cream/lotion/moisturizer'}).replace({'crème': 'cream/lotion/moisturizer'}).replace({'oil': 'cream/lotion/moisturizer'}).replace({'moisturising': 'cream/lotion/moisturizer'})
```

```python
df["product_type"]=df['product_type'].replace({'facial wash': 'cleanser'}).replace({'wash': 'cleanser'}).replace({'cleansing': 'cleanser'}).replace({'balm': 'cleanser'}).replace({'micellar': 'cleanser'})
```

```python
df["product_type"]=df['product_type'].replace({'exfoliator': 'exfoliant'}).replace({'scrub': 'exfoliant'})
```

```python
df["product_type"]=df['product_type'].replace({'essence': 'toner/essence/emulsion'}).replace({'emulsion': 'toner/essence/emulsion'}).replace({'toner': 'toner/essence/emulsion'}).replace({'tonic': 'toner/essence/emulsion'}).replace({'essence': 'toner/essence/emulsion'}).replace({'facial mist': 'toner/essence/emulsion'}).replace({'face mist': 'toner/essence/emulsion'}).replace({'mist': 'toner/essence/emulsion'}).replace({'booster': 'toner/essence/emulsion'})
```

```python
df["product_type"]=df['product_type'].replace({'retinol': 'treatment'})
```

```python
df["product_type"]=df['product_type'].replace({'masque': 'mask'})
```
